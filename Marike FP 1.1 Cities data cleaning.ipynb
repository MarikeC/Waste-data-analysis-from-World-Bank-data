```python
cities_df.replace({'LIC': 'LowIncome', 'LMC': 'LowMiddleIncome','UMC':'MiddleUpperIncome','HIC':'HighIncome'},inplace=True)
```

```python
Cities_null_perc_X_region_income = pd.DataFrame()
Cities_null_perc_X_region_income['World']=cities_null_perc['Missing data (%)']

for x in cities_df.region_id.unique():
    null_perc_city_region = show_null(cities_df.loc[cities_df.region_id == x])
    Cities_null_perc_X_region_income[x]=round(null_perc_city_region['Missing data (%)'],2)
    
for x in cities_df.income_id.unique():
    null_perc_city_income = show_null(cities_df.loc[cities_df.income_id == x])
    Cities_null_perc_X_region_income[x]=round(null_perc_city_income['Missing data (%)'],2)

Cities_null_perc_X_region_income
```

We can see that some variables have a very high percentage of missing information. **We decide to keep only variables with less than 50% of missing information except for variables with 50 to 70% missing information but for which there is a world area (defined geographically or by income) where this variable has less than 30% missing information.** 

Indeed we know that some variables will only be informed in areas where waste issue is better dealt with.

```python
Notallregions_Income_HighNaNPerc=[]

for x in Cities_null_perc_X_region_income.index:
    
    if Cities_null_perc_X_region_income.at[x,'World']<70 and Cities_null_perc_X_region_income.at[x,'World']>50:
        
        for y in cities_df.region_id.unique():
            if Cities_null_perc_X_region_income.at[x,y]<30:
                Notallregions_Income_HighNaNPerc.append(x)
                
        for z in cities_df.income_id.unique():
            if Cities_null_perc_X_region_income.at[x,z]<30:
                Notallregions_Income_HighNaNPerc.append(x)
                
Notallregions_Income_HighNaNPerc
```

```python
Notallregions_Income_HighNaNPerc_df = pd.DataFrame({'Variable to keep':Notallregions_Income_HighNaNPerc})
```

```python
Notallregions_Income_HighNaNPerc_df['Variable to keep'].unique()
```

```python
ColToDrop = []
for x in Cities_null_perc_X_region_income.index:
    if Cities_null_perc_X_region_income.at[x,'World']>50:
        if x not in Notallregions_Income_HighNaNPerc_df['Variable to keep'].unique():
            ColToDrop.append(x)

ColToDrop
```

```python
cities_df_1 = cities_df.drop(ColToDrop, axis=1)
cities_df_1
```
