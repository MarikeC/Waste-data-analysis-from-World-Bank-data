Countries have the variables *region_id* and *income_id* in commmon with cities. So again, before dropping massively information, we will check if the missing information is so important for each sub geographical or economic area.

```python
Countries_null_perc_X_region_income = pd.DataFrame()

Countries_null_perc_X_region_income['World']=countries_null_perc['Missing data (%)']

for x in countries_df.region_id.unique():
    null_perc_country_region = show_null(countries_df.loc[countries_df.region_id == x])
    Countries_null_perc_X_region_income[x]= round(null_perc_country_region['Missing data (%)'],2)

for x in countries_df.income_id.unique():
    null_perc_country_income = show_null(countries_df.loc[countries_df.income_id == x])
    Countries_null_perc_X_region_income[x]= round(null_perc_country_income['Missing data (%)'],2)    

Countries_null_perc_X_region_income
```

AS with cities dataset, we can see that some variables have a very high percentage of missing information for all regions/income type. We keep same criteria as previously with cities: we decide to keep only variables with less than 50% of missing information except for variables with 50 to 70% missing information but for which there is a world area (defined geographically or by income) where this variable has less than 30% missing information.


```python
Countries_Notallregions_Income_HighNaNPerc=[]

for x in Countries_null_perc_X_region_income.index:
    
    if Countries_null_perc_X_region_income.at[x,'World']<70 and Countries_null_perc_X_region_income.at[x,'World']>50:
        
        for y in countries_df.region_id.unique():
            if Countries_null_perc_X_region_income.at[x,y]<30:
                Countries_Notallregions_Income_HighNaNPerc.append(x)
                
        for z in countries_df.income_id.unique():
            if Countries_null_perc_X_region_income.at[x,z]<30:
                Countries_Notallregions_Income_HighNaNPerc.append(x)
                
Countries_Notallregions_Income_HighNaNPerc
```

In order not to loose the information in the rows corresponding to missing gdp, since income_id and gdp are related, when gdp is missing we want to replace it by the mean of gdp for the countries with same income_id.
Nevertheless GDP value is also related to population so in order to realize this operation we first create a new variable : GDP by inhabitant (gdp_inhab) which will not depend on country size.

```python
countries_df_3['gdp_inhab']=round(countries_df_3['gdp']/countries_df_3['population'],2)
```

```python
mean_gdp_inhab_by_income_dict = {}
for x in countries_df_3.income_id.unique():
    mean_gdp_inhab_by_income_dict.update({x:countries_df_3.gdp_inhab.loc[countries_df_3.income_id == x].mean()}) 

mean_gdp_inhab_by_income_dict
```

```python
for x in countries_df_3.income_id.unique():
    countries_df_3['gdp_inhab'].loc[countries_df_3.income_id == x]=countries_df_3['gdp_inhab'].loc[countries_df_3.income_id == x].fillna(mean_gdp_inhab_by_income_dict[x])
```
